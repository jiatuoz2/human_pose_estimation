```python
import numpy as np
from imageio import imread
import matplotlib.pyplot as plt
import math
from scipy import signal
%matplotlib inline

#Hough Transform Function

def rgb2gray(rgb):
    grayimg = np.dot(rgb, [0.2989, 0.5870, 0.1140]) # element-wise dot product
    return grayimg

#Canny Edge Detection Function

def Canny(img):
    gaussian_k = np.array([[2, 4, 5, 4, 2],
                          [4, 9, 12, 9, 4], 
                          [5, 12, 15, 12, 5],
                          [4, 9, 12, 9, 4],
                          [2, 4, 5, 4, 2]]) / 159

    img = signal.convolve2d(gaussian_k, img)

    kx = np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]])
    ky = np.array([[-1, -2, -1],
                   [0, 0, 0],
                   [1, 2, 1]])

    Gx = signal.convolve2d(kx, img)
    Gy = signal.convolve2d(ky, img)

    G_mag = np.sqrt(Gx**2 + Gy**2)

    G_dir = np.arctan2(Gy, Gx) * 180 / np.pi
    
    return G_mag, G_dir 

def NMS(G_mag, G_dir):
    n1, n2 = G_mag.shape
    img_NMS = np.zeros((n1, n2))
    
    angles = np.round(G_dir / 45) % 4 # 4 lines 
    
    for i in range(n1):
        for j in range(n2):
            angle = angles[i, j]
            if angle == 0:
                if i > 0 and G_mag[i - 1, j] > G_mag[i, j]:
                    img_NMS[i, j] = 0
                elif i < n1 - 1 and G_mag[i + 1, j] > G_mag[i, j]:
                    img_NMS[i, j] = 0
                else:
                    img_NMS[i, j] = G_mag[i, j]
            elif angle == 3:
                if i > 0 and j < n2 - 1 and G_mag[i - 1, j + 1] > G_mag[i, j]:
                    img_NMS[i, j] = 0
                elif i < n1 - 1 and j > 0 and G_mag[i + 1, j - 1] > G_mag[i, j]:
                    img_NMS[i, j] = 0
                else:
                    img_NMS[i, j] = G_mag[i, j]
            elif angle == 2:
                if j > 0 and G_mag[i, j - 1] > G_mag[i, j]:
                    img_NMS[i, j] = 0
                elif j < n2 - 1 and G_mag[i, j + 1] > G_mag[i, j]:
                    img_NMS[i, j] = 0
                else:
                    img_NMS[i, j] = G_mag[i, j]
            elif angle == 1:
                if i > 0 and j > 0 and G_mag[i - 1, j - 1] > G_mag[i, j]:
                    img_NMS[i, j] = 0
                elif i < n1 - 1 and j < n2 - 1 and G_mag[i + 1, j + 1] > G_mag[i, j]:
                    img_NMS[i, j] = 0
                else:
                    img_NMS[i, j] = G_mag[i, j]
    
    return img_NMS

def Hough_transform(img):
    n1, n2 = img.shape
    D = int(np.round(np.sqrt(n1**2 + n2**2))) 
    accumulator_cells = np.zeros((181, 2 * D + 1))
    
    plt.figure()
    plt.xlabel("theta")
    plt.ylabel("rho")
    plt.xlim(0, 180)
    plt.ylim(-D, D)
    plt.title("HT graph of test image")
    
    x_indices, y_indices = np.nonzero(img)
    
    for i in range(len(x_indices)):
        x = x_indices[i]
        y = y_indices[i]
        
        theta = np.linspace(0, 180, 181)
        rho = np.cos(np.radians(theta)) * x + np.sin(np.radians(theta)) * y
        theta = theta[(rho < D + 1) & (rho >= -D)]
        rho = rho[(rho < D + 1) & (rho >= -D)]


        plt.plot(theta, rho)
        rho = np.round(rho) + D
        for k in range(len(theta)):
            accumulator_cells[int(theta[k]), int(rho[k])] += 1
    
    max_HT = np.max(accumulator_cells)
    theta, rho = np.where(accumulator_cells > 0.75 * max_HT)
    
    rho -= D
    
    print(theta, rho)
    return (rho, theta)


def plot_detected_line(rho, theta, img):
    n1, n2 = img.shape
    plt.figure()
    
    for i in range(len(rho)):
        theta_radian = np.radians(theta[i])
        
        # find x intercept
        x0 = np.round(rho[i] / np.cos(theta_radian))
        x1 = np.round((rho[i] - (n2 - 1) * np.sin(theta_radian)) / (np.cos(theta_radian)))
        
        # find y intercept 
        y0 = np.round(rho[i] / np.sin(theta_radian))
        y1 = np.round((rho[i] - (n1 - 1) * np.cos(theta_radian)) / (np.sin(theta_radian)))
                
        if x0 < n1 and x0 >= 0:
            x0 = int(x0)
            if x1 < n1 and x1 >= 0:
                x1 = int(x1)
                plt.plot([0, n2 - 1], [x0, x1], color="green", linewidth=3)
            elif y0 < n2 and y0 >= 0:
                y0 = int(y0)
                plt.plot([0, y0], [x0, 0], color="green", linewidth=3)
            elif y1 < n2 and y1 >= 0:
                y1 = int(y1)
                plt.plot([0, y0], [x0, n1 - 1], color="green", linewidth=3)
        elif x1 >= 0 and x1 < n1:
            x1 = int(x1)
            if y0 < n2 and y0 >= 0:
                y0 = int(y0)
                plt.plot([n2 - 1, y0], [x1, 0], color="green", linewidth=3)
            elif y1 < n2 and y1 >= 0:
                y1 = int(y1)
                plt.plot([n2 - 1, y1], [x1, n1 - 1], color="green", linewidth=3)
        elif y0 < n2 and y0 >= 0:
            y0 = int(y0)
            y1 = int(y1)
            plt.plot([y0, y1], [0, n1 - 1], color="green", linewidth=3)
    plt.imshow(img, cmap="gray")
        
```

```python
# create your 11x11 test image and load "lane.png"
test = np.zeros([11, 11])
test[0, 10] = 1
test[5, 5] = 1
test[10, 0] = 1
test[10, 10] = 1
for i in range(11):
    test[i, i] = 255
plt.imshow(test,interpolation='nearest', aspect='equal',cmap='gray')

#Sample call and plot
rho, theta = Hough_transform(test)
plot_detected_line(rho, theta, test)
```
