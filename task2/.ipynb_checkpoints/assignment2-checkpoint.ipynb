## Problem 1 Image shift
Shifting an image x of size (n1, n2) in a direction (k, l) consists in creating a new image xshifted of size
(n1, n2) such that

<img src="markdown/fig1.png">

In practice, boundary conditions should be considered for pixels (i, j) such that (i + k, j + l) not equal to [0, n1-1] x [0, n2-1]. 

A typical example is to consider periodical boundary conditions such that

<img src="markdown/fig2.png">

Create in imshift function implementing the shifting of an image x in periodical boundary, such as the following image(b) Shifted in the direction (k,l) by (+100,-50):
<img src="markdown/fig3.png">

Hint: First write it using loops, and next try to get rid of the loops.

```python
import numpy as np
from imageio import imread
import matplotlib.pyplot as plt

def imshift(x, k, l):
    n1, n2 = x.shape
    xshifted = np.zeros((n1, n2))
    temp_shifted = np.zeros((n1, n2))
    
    real_k = k % n1
    real_l = l % n2
    
    # shift in x
    if real_k != 0:
        temp_shifted[real_k:, :] = x[:(n1 - real_k) % n1, :]
        temp_shifted[:real_k, :] = x[(n1 - real_k) % n1:, :]
    else:
        temp_shifted = x.copy()
    
    # shift in y
    if real_l != 0:
        xshifted[:, real_l:] = temp_shifted[:, :(n2 - real_l) % n2]
        xshifted[:, :real_l] = temp_shifted[:, (n2 - real_l) % n2:]
    else:
        xshifted = temp_shifted 
        
    return xshifted

#Sample call and Plotting code
#lake.png and "windmill.png"
lake = imread("lake.png")
windmill = imread("windmill.png")

plt.subplot(2, 2, 1)
plt.imshow(lake)
plt.subplot(2, 2, 2)
plt.imshow(windmill)

lake_shifted = imshift(lake, 100, -50)
windmill_shifted = imshift(windmill, 100, -50)

plt.subplot(2, 2, 3)
plt.imshow(lake_shifted)
plt.subplot(2, 2, 4)
plt.imshow(windmill_shifted)
```

```python
# check linearity
a = 1
b = 2
k = 100
l = -50

result1 = imshift(a * lake + b * windmill, k, l)
result2 = a * imshift(lake, k, l) + b * imshift(windmill, k, l)

plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(result1)
plt.subplot(1, 2, 2)
plt.imshow(result2)

# the result is not one-to-one for the first shift

# shift back
windmill_recovered = imshift(windmill_shifted, -k, -l)
lake_recovered = imshift(lake_shifted, -k, -l)

plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(lake_recovered)
plt.subplot(1, 2, 2)
plt.imshow(windmill_recovered)

# the result is one-to-one after shifting back
```

The spatial method intend to create a stacked-up image, which is a 3D matrix with dimension [ image_height x image_width x total_kernel_number]. The final code should read with only two loops. 

The idea is to switch the inner loops with the outer loops, and then make use of imshift to eliminate the need to use the outter loop to go over the entire image, but use shift function to shift the image along the inner loop and store the each result into your 3D matrix. With the stacked-up matrix, you can perform convolution calculation with ease. on the 3D matrix.

```python
import numpy as np
from imageio import imread
import matplotlib.pyplot as plt
import time

def imkernel(tau, s1, s2):
    '''
    The kernel (window) function already, you can use that function to generate your kernel. 

    Note: the function is slightly difference than just a matrix. As you can see it returned a 
    lambda function object. You need to assign location of the kernel, then it will return 
    specified value in that location of the kernel.
    
    For example, we want a 3x3 window, (note: in this function, we said the center point to be 
    location (0,0), so s1 is the absolution distance to the center point, for example: s1 means 
    from -1 to 1): 

    nu = imkernel(tau,s1,s1);   #<------- generated a 3x3 window funtion
    nu(-1,-1)  #<--------- this will return the top left corner value of the kernel
    nu(0,0)  #<--------- this will return the center value of the kernel

    '''
    # 2x + 2y
    #equation = lambda x, y: 2x + 2y
    #equation(1 , 2)
    
    w = lambda i,j:np.exp(-(i**2+j**2)/(2*tau**2))
    # normalization
    i,j=np.mgrid[-s1:s1,-s2:s2]
    Z = np.sum(w(i,j))
    nu = lambda i,j: w(i,j)/Z*(np.absolute(i)<= s1&np.absolute(j)<=s2)
    #Note: The return value is a lambda function, which require (i,j) input.
    #nu = w(i,j)/Z
    return nu


# Create imconvolve_naive function,
def imconvolve_naive(im, nu, s1, s2):
    (n1,n2)=im.shape
    xconv = np.zeros((n1,n2))
                    
    for i in range(s1, n1 - s1):
        for j in range(s2, n2 - s2):
            for k in range(-s1, s1 + 1):
                for l in range(-s2, s2 + 1):
                    xconv[i, j] += im[i - k, j - l] * nu(k, l)
    
    return xconv


#Create imconvolve_spatial function
def imconvolve_spatial(im,nu, s1,s2):
    (n1, n2) = im.shape
    xconv = np.zeros((n1, n2))
    
    for i in range(s1 * 2 + 1):
        for j in range(s2 * 2 + 1):
            temp_im = imshift(im, -i, -j)
            xconv += temp_im * nu(i - s1, j - s2)

    return xconv

#Sample call and Plotting code
tau = 1
s1 = 2
s2 = 2

nu = imkernel(tau, s1, s2)
windmill_conv = imconvolve_naive(windmill, nu, s1, s2)
windmill_conv_spatial = imconvolve_spatial(windmill, nu, s1, s2)

plt.figure()
plt.imshow(windmill)
plt.figure()
plt.imshow(windmill_conv)
plt.figure()
plt.imshow(windmill_conv_spatial)

def test_imconvolve():
    start_naive = time.time()
    windmill_naive = imconvolve_naive(windmill, nu, s1, s2)
    end_naive = time.time()
    
    start_spatial = time.time()
    windmill_spatial = imconvolve_spatial(windmill, nu, s1, s2)
    end_spatial = time.time()
    
    print("The execution time for naive convolution is ", end_naive - start_naive)
    print("The execution time for spatial convolution is ", end_spatial - start_spatial)
    
    difference = windmill_spatial - windmill_naive
    plt.figure()
    plt.imshow(difference)
    
test_imconvolve()
```

## Problem 3: Order-statistic filtering

Order-statistic filters (OSF) are local filters that are only based on the ranking of pixel values inside a sliding window.
1. Create in imstack(img,s1,s2) function that creates a stack xstack of size n1 n2 s, which s = (2s1 +1)(2s2 +1) from the n1 n2 image x, such that xstack(i,j,:) contains all the values of x in the neighborhood (s1, s1)  (s2, s2). This function should take into account the four possible boundary conditions.

    Hint: you can use imshift, which we implemented in assignment 1, and only two loops for s1 <= k <= s1 and s2<= l<= s2.
    
2. Create in imosf() function
     function imosf(x, type, s1, s2) that implements order-statistic filters, returns xosf. imosf should first call imstack, next sort the entries of the stack with respect to the third dimension, and create the suitable output xosf according to the string type as follows:
     
     'median': select the median value,
    
     'erode': select the min value,
    
     'dilate': select the max value,
    
     'trimmed': take the mean after excluding at least 25% of the extreme values on each side.
    
3. Create in imopening() and imclosing() function that performs the opening and closing by the means of OSF filters.
    <img src="markdown/fig6.png">

4. Load castle.png. Write a script to test imosf() that loads the image x = castle and create a corrupted version of image x with 10% of impulse noise (salt and pepper)

Apply your OSF filters and zoom on the results to check that your results are consistent with the following ones:
    <img src="markdown/fig7.png">

Note: Typo on image, (d) -- opening , (e) -- closing


```python
#Function
def imstack(img, s1, s2):
    (n1, n2) = img.shape
    s = (2 * s1 + 1) * (2 * s2 + 1)
    
    xstack = np.zeros((n1, n2, s))
    
    for i in range(-s1, s1 + 1):
        for j in range(-s2, s2 + 1):
            im_shifted = imshift(img, -i, -j)
            if -i > 0:
                im_shifted[:-i, :] = 0
            elif -i < 0:
                im_shifted[-i:, :] = 0
            if -j > 0:
                im_shifted[:, :-j] = 0
            elif -j < 0:
                im_shifted[:, -j:] = 0
            
            xstack[:, :, (i + 1) * (2 * s2 + 1) + j + 1] = im_shifted
        
    return xstack

def imosf(x, type, s1, s2):
    xstack = imstack(x, s1, s2)
    n1, n2, s = xstack.shape
    
    xosf = np.zeros((n1, n2))
    
    for i in range(n1):
        for j in range(n2):
            neighborhood = xstack[i, j, :]
            
            if type == "median":
                xosf[i, j] = np.median(neighborhood)
            elif type == "erode":
                xosf[i, j] = np.min(neighborhood)
            elif type == "dilate":
                xosf[i, j] = np.max(neighborhood)
            elif type == "trimmed":
                sorted_neighborhood = np.sort(neighborhood)
                trimmed_length = int(s * 0.25)
                xosf[i, j] = np.mean(sorted_neighborhood[trimmed_length:(s - trimmed_length)])
    return xosf

def imopening(x, s1, s2):
    result = imosf(x, "erode", s1, s2)
    result = imosf(result, "dilate", s1, s2)
    return result

def imclosing(x, s1, s2):
    result = imosf(x, "dilate", s1, s2)
    result = imosf(result, "erode", s1, s2)
    return result

#Import image here
# Sample call
# castle.png
castle = imread("castle.png")
plt.figure()
plt.imshow(castle)

castle_noisy = castle.copy()
n1, n2 = castle.shape
num_pixel = n1 * n2
num_noisy_pixel = int(num_pixel * 0.1)
s1 = 2
s2 = 2

rand_x_coord = np.random.randint(0, n1, num_noisy_pixel) 
rand_y_coord = np.random.randint(0, n2, num_noisy_pixel)

for i in range(num_noisy_pixel):
    rand_pixel = np.random.randint(0, 256)
    castle_noisy[rand_x_coord[i], rand_y_coord[i]] = rand_pixel

plt.figure()
plt.imshow(castle_noisy)

castle_median = imosf(castle, "median", s1, s2)
castle_trimmed = imosf(castle, "trimmed", s1, s2)
castle_closing = imclosing(castle, s1, s2)
castle_opening = imopening(castle, s1, s2)

plt.figure()
plt.imshow(castle_median)
plt.figure()
plt.imshow(castle_trimmed)
plt.figure()
plt.imshow(castle_opening)
plt.figure()
plt.imshow(castle_closing)
```

(iii) Non-maximum Suppression (NMS): 

The purpose of this step is to convert the thick edges in the gradient magnitude image to sharp edges. This is done by preserving all local maxima in the gradient image, and deleting everything else. This is carried out by recursively performing the following steps for each pixel in the gradient image:

 Round the gradient direction  to nearest 45, corresponding to the use of an 8-connected neighbourhood.

 Compare the edge strength of the current pixel with the edge strength of the pixel in the positive and negative gradient direction i.e. if the gradient direction is north ( = 90), then compare with the pixels to the north and south.

 If the edge strength of the current pixel is largest; preserve the value of the edge strength. If not, suppress (remove) the value.

(iv) Thresholding: 

The edge-pixels remaining after the NMS step are (still) marked with their strength. Many of these will probably be true edges in the image, but some may be caused by noise or color variations. The simplest way to remove these would be to use a threshold, so that only edges stronger that a certain value would be preserved.
